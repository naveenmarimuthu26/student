```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import StandardScaler

# ---------------------------
# FIXED DATA
# ---------------------------
x = np.array([1, 5, 7, 8, 9])
y = np.array([1, 2, 3, 4, 5])    # <-- You forgot to define y

print(x.shape)
print(y.shape)
print(x)
print(y)

x = x.reshape(-1, 1)
print(x.shape)
print(x)

# ---------------------------
# LINEAR REGRESSION
# ---------------------------
model = LinearRegression()
model.fit(x, y)

print(model.predict([[8]]))

y_pred = model.predict(x)

plt.scatter(x, y, color="r", marker="o")
plt.plot(x, y_pred, color="g", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

print(model.coef_)
print(model.intercept_)

# ---------------------------
# SGD REGRESSOR + STANDARDIZATION
# ---------------------------
scaler = StandardScaler()
x_stand = scaler.fit_transform(x.astype(float))

sgdr = SGDRegressor(penalty="l2", alpha=0.15, max_iter=1000)  # FIXED
sgdr.fit(x_stand, y)

print(sgdr.predict(scaler.transform([[10]])))   # <-- Must scale input!

y_pred2 = sgdr.predict(x_stand)

plt.scatter(x_stand, y, color="y", marker="o")
plt.plot(x_stand, y_pred2, color="b", linewidth=2)
plt.xlabel("x_normalized")
plt.ylabel("y")
plt.show()

print(sgdr.coef_)
print(sgdr.intercept_)
```
