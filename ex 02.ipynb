```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
Create Numpy arrays
x = np.array([1,2,4,3,5])
y = np.array([1,3,4,3,7])
Print shapes
print(x.shape)
print(y.shape)
print(x)
print(y)
Convert x into nd array
x = x.reshape(-1,1)
print(x.shape)
print(x)
Create the model
model = LinearRegression()
model.fit(x, y)
Predict value for unknown x
#predict y for an unknown x=8
print(model.predict([[8]]))

P24CS2P3 – Machine Learning Lab

Department of Computer Science (Shift – I), Bishop Heber College, Trichy -17 | 7
#predict all x values
y_pred = model.predict(x)
Draw scatter plot
plt.scatter(x, y, color='r', marker='o')
plt.plot(x, y_pred, color='b', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
# Let us print weight and intercept values
print(model.coef_)
# Print intercept value
print(model.intercept_)
Multiple Linear Regression with a simple example
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_stand = scaler.fit_transform(x.astype(float))
Create stochostic gradient descent regressor model
sgdr = SGDRegressor(penalty='l2', alpha=0.15, max_iter=100)
Train the model
sgdr.fit(x_stand, y)
Predict value for the given input
print(sgdr.predict([[10]]))
Predict values for X
y_pred2 = sgdr.predict(x_stand)
Draw scatter plot
plt.scatter(x_stand, y, color='g', marker='o')
plt.plot(x_stand, y_pred2, color='b', linewidth=2)
plt.xlabel('x_normalized')
plt.ylabel('y')
plt.show()

P24CS2P3 – Machine Learning Lab

Department of Computer Science (Shift – I), Bishop Heber College, Trichy -17 | 8
print(sgdr.coef_)
print(sgdr.intercept_)
```
